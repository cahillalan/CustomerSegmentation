#  Customer Segmentation - Data Cleaning

##  Objective
The goal of this notebook is to clean and prepare customer data for segmentation using clustering techniques.

## Business Context
Customer segmentation relies on identifying patterns in customer behaviour.

To ensure meaningful clusters:
- Data must be accurate
- Missing values must be handled
- Features must reflect real customer behaviour

Poor data quality would lead to unreliable customer segments.

```python
import pandas as pd
import numpy as np

df = pd.read_excel('../Data/raw/Online_Retail.xlsx')
```

## Data Cleaning

Data cleaning is performed to ensure the dataset is suitable for segmentation.

This includes:
- Handling missing values
- Correcting data types
- Removing invalid or inconsistent records

### Why This Matters for Clustering

Clustering algorithms rely on patterns in the data.

If the data contains:
- Missing values
- Incorrect formats
- Noise

Then the resulting customer segments will be unreliable.

Clean data ensures that clusters reflect real customer behaviour rather than data issues.

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.isnull().sum()
```

```python
df = df.dropna(subset =["customerid"])
```

```python
df["description"]= df["description"].fillna('Unknown')
```

```python
df['customerid'] =df['customerid'].astype(int)
df['invoicedate']= pd.to_datetime(df['invoicedate'])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df =df[~df['invoiceno'].astype(str).str.startswith('C')]
```

```python
df = df[df['quantity'] >0]
df = df[df['unitprice']>0]
```

```python
df['total_price']=df['quantity'] * df['unitprice']
```

```python
df.describe()
```

```python
df.to_csv("../Data/processed/online_retail_cleaned.csv", index = False)
```
